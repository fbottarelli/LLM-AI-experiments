```python
import os
import anthropic
from IPython.display import display, Markdown

# Carica le variabili d'ambiente da .env
from dotenv import load_dotenv
load_dotenv()

# Inizializza il client Anthropic
client = anthropic.Anthropic()

# Funzione per eseguire l'analisi
def analyze_cloudera_reports(old_report, new_report):
    prompt = f"""You are an AI assistant tasked with analyzing and comparing two Cloudera cluster reports: an old report and a new report. Your goal is to identify and highlight the differences between these reports, focusing on changes in services, potential problems, and other notable variations. 

Here is the old report:
<old_report>
{old_report}
</old_report>

Here is the new report:
<new_report>
{new_report}
</new_report>

Analyze both reports and identify the key differences between them. Pay special attention to:
1. Changes in the status or configuration of Cloudera services
2. New or resolved issues
3. Performance variations
4. Any other significant changes between the old and new reports

Provide a concise summary of your findings, highlighting the most important differences. Be brief and to the point. Your response should be structured as follows:

<analysis>
1. [First key difference]: Brief explanation
2. [Second key difference]: Brief explanation
3. [Third key difference]: Brief explanation
...
</analysis>

<summary>
Provide a brief overall summary of the changes, focusing on their potential impact on the Cloudera cluster's performance and stability.
</summary>

Remember to:
- Focus only on significant changes
- Use technical terms related to Cloudera and its services when appropriate
- Keep your explanations concise
- Limit your response to the most important 3-5 differences, unless there are more critical changes that must be mentioned
- answer in italian

If there are no significant differences between the old and new reports, state this clearly in your analysis.
"""

    response = client.messages.create(
        model="claude-3-5-sonnet-20240620",
        max_tokens=1000,
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )
    
    return response.content[0].text
```

```python
# Input dei report
old_report = input("Inserisci il vecchio report Cloudera: ")
new_report = input("Inserisci il nuovo report Cloudera: ")

# Esegui l'analisi
analysis = analyze_cloudera_reports(old_report, new_report)

# Visualizza il risultato
display(Markdown(analysis))
```
