```python
import os
import anthropic
from IPython.display import display, Markdown
import base64

# Carica le variabili d'ambiente da .env
from dotenv import load_dotenv
load_dotenv()

# Inizializza il client Anthropic
client = anthropic.Anthropic()
import base64

# Funzione per eseguire l'analisi
def extract_blood_data(image_path):
    prompt = f"""Describe what you see in the following blood test results
    """

    # Apre il file immagine e lo legge in modalità binaria
    with open(image_path, "rb") as image_file:
        image1_data = base64.b64encode(image_file.read()).decode("utf-8")

    response = client.messages.create(
        model="claude-3-5-sonnet-20240620",
        max_tokens=1000,
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "image",
                        "source": {
                            "type": "base64",
                            "media_type": "image/jpeg",
                            "data": image1_data,
                        },
                    },
                    {
                        "type": "text",
                        "text": "Extract the blood test data from the image and the date of the test. The data must be a CSV format with the following columns: 'Test Name', 'Result', 'Reference Range', 'Units', 'Date'. The date must be in the format 'YYYY-MM-DD'., Just write the csv data here:",
                    }
                ],
            }
        ],
    )

    return response.content[0].text
```

```python
image_path = "data/tamara_images/photo_5926902182848873542_y.jpg"
result = extract_blood_data(image_path)
print(result)
# save it to a file csv
with open("data/tamara_images/blood_test.csv", "w") as f:
    f.write(result)
print("File saved to data/tamara_images/blood_test.csv")

# now do it for all jpgs in data/tamara_images
for file in os.listdir("data/tamara_images"):
    if file.endswith(".jpg"):
        image_path = os.path.join("data/tamara_images", file)
        result = extract_blood_data(image_path)
        print(result)
        # save it to a file csv
        with open(f"data/tamara_images/{file}.csv", "w") as f:
            f.write(result)
        print(f"File saved to data/tamara_images/{file}.csv")
```
